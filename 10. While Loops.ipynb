# `while` Loops

## Iteration with `while` loops

**Syntax is** 

`while` loops have the following syntax:

```
while condition:
    statement(s)
```

Statements in the code block under `while` are executed repeatedly as long as the `condition` evaluates to `True`. Generally, one of the statements under `while` makes some change to a variable that causes the condition to evaluate to `False` after a certain number of iterations.








**Example 1.** Display first 10 numbers 

````python
count=1
print('The count is:')
while (count <=10):
    print(count)
    count=count+1
print('Good Day!')    
````

**Note:** The block here, consisting of the print and increment statements, is executed repeatedly until count is no longer less than 9. With each iteration, the current value of the index count is displayed and then increased by 1.

**Example 6.** Let's try to calculate the factorial of `10` using a `while` loop. The factorial of a number `n` is the product (multiplication) of all the numbers from `1` to `n`, i.e., `1*2*3*...*(n-2)*(n-1)*n`.

````python
result=1 
i=1 
n=int(input('Factorial of = '))#since it takes user input you can find factorial of other numbers as well
while i<=n:
    result = result*i
    i=i+1
print('The factorial of {} is {}'.format(n,result))    
    
````

Here's how the above code works:

* We initialize two variables, `result` and, `i`. `result` will contain the final outcome. And `i` is used to keep track of the next number to be multiplied with `result`. Both are initialized to 1 (can you explain why?)

* The condition `i <= 10` holds true (since `i` is initially `1`), so the `while` block is executed.

* The `result` is updated to `result * i`, `i` is increased by `1` and it now has the value `2`.

* At this point, the condition `i <= 11` is evaluated again. Since it continues to hold true, `result` is again updated to `result * i`, and `i` is increased to `3`.

* This process is repeated till the condition becomes false, which happens when `i` holds the value `11`. Once the condition evaluates to `False`, the execution of the loop ends, and the `print` statement below it is executed. 

Can you see why `result` contains the value of the factorial of 10 at the end? If not, try adding `print` statements inside the `while` block to print `result` and `i` in each iteration.

### Using `else` Statement with `While` Loop

Python supports to have an `else` statement associated with a loop statement.

If the `else` statement is used with a `while` loop, the `else` statement is executed when the condition becomes false.


**Example 7.** The following example illustrates the combination of an else statement with a while statement that prints a number as long as it is less than 5, otherwise else statement gets executed.


````python
count=0
while count<5:
    print(count,'is less than 5')
    count=count+1
else:
    print(count,'is not less than 5')
````

## Loop Control Statements

Loop control statements change execution from its normal sequence. When execution leaves a scope, all automatic objects that were created in that scope are destroyed. 
Python supports the following control statements. 

### `break` statement

You can use the `break` statement within the loop's body to immediately stop the execution and *break* out of the loop (even if the condition provided to `while` still holds true).


**Syntax**

The syntax for a break statement in Python is as follows −

`break`













**Example 8.** Demonstrate `break` statement

````python
var=10
while var>0:
    print('current variable value:',var)
    var=var-1
    if var==5:
        break
print('Good Morning')
#Breaks the loop when var=5
````

````python
i=1
result=1
while i<=10:
    result*=i #result=result*1
    if i==4:
        print('Magic number 4 reached: stop execution..')
        break
    i+=1 #i=i+1
print('i:',i)
print('result:',result)
#insted of else we used break in here
````

### `continue` statement

It returns the control to the beginning of the while loop.. The `continue` statement rejects all the remaining statements in the current iteration of the loop and moves the control back to the top of the loop.

The `continue` statement can be used in both while and for loops.








As you can see from above example that, the value of `i` at the end of execution is 4. This example also shows how you can use an `if` statement within a `while` loop.

Sometimes you may not want to end the loop entirely, but simply skip the remaining statements in the loop and *continue* to the next loop. You can do this using the `continue` statement.

````python
i=1
result=1
while i<10:
    i+=1
    if i%2==0:
        print('skipping {}'.format(i))
        continue
    print('Multiplication with {}'.format(i))
    result*=i
print('i:',i)
print('result:',result)
#only mutiplies when not even
````

In the example above, the statement `result = result * i` inside the loop is skipped when `i` is even, as indicated by the messages printed during execution.
